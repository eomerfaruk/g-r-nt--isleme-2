```python
import numpy as np 
import cv2
from matplotlib import pyplot as plt 
```

```python
zero_matris = np.zeros(shape=(512,512,3) , dtype = np.int16)
```

```python
plt.imshow(zero_matris)
```

```python
cv2.putText(zero_matris , text = "omererdem",org = (10,500) , fontFace=cv2.FONT_HERSHEY_SIMPLEX , fontScale = 2 ,color= (255,255,255) , thickness = 2 )
```

```python
plt.imshow(zero_matris)
```

```python
noktalar = np.array([[100,400],[200,200],[400,300],[200,400]]  ,dtype = np.int32 )
```

```python
cv2.polylines(zero_matris , [noktalar],isClosed=True , color=(255,0,0),thickness=2)
```
