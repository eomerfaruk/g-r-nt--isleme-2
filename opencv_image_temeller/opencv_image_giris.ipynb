```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
img = cv2.imread("data\\kopek.jpg")
plt.imshow(img)
```

```python
img.shape
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR -> RGB
```

```python
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY) #BGR -> GRAY , gri format tek kanall matris
```

```python
plt.imshow(img_gray)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray,cmap = "gray" )
```

```python
new_img = cv2.resize(img ,(1800,500))
```

```python
plt.imshow(new_img)
```
