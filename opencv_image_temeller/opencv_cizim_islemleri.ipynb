```python
import cv2 
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
zero_matris = np.zeros(shape=(512,512,3), dtype=np.int16 )  
```

```python
plt.imshow(zero_matris)
```

```python
cv2.rectangle(zero_matris , (300,100) , (400,300) , (0,255,0) , 2) # görüntümüzün üstüne cerceve cizer
```

```python
plt.imshow(zero_matris)
```

```python
cv2.circle(zero_matris , center = (100,100),radius = 50 , color = (0,0,255) , thickness=2)
```

```python
plt.imshow(zero_matris)
```

```python
cv2.circle(zero_matris , center = (100,100),radius = 50 , color = (0,0,255) , thickness=-1)
```

```python
plt.imshow(zero_matris)
```

```python
cv2.line(zero_matris , pt1 = (0,0) ,pt2 = (512,512) ,color=(0,0,255) , thickness = 2)
```

```python
plt.imshow(zero_matris)
```
