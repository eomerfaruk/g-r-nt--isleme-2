```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
img = cv2.imread("DATA\\internal_external.png" , 0)
```

```python
plt.imshow(img ,cmap="gray")
```

```python
img.shape
```

```python
contours ,hierarchy = cv2.findContours(img ,cv2.RETR_CCOMP , cv2.CHAIN_APPROX_SIMPLE) #RETR_CCOMP  hen external hemde tree contourlarını alır
```

```python
external_contours = np.zeros(img.shape)
```

```python
plt.imshow(external_contours)
```

```python
for i in range(len(contours)):
    if hierarchy[0][i][3] == -1: #en dış kısmındaki hierarchy external olan contourlar ise
        cv2.drawContours(external_contours , contours , i ,255 ,-1)
```

```python
plt.imshow(external_contours)
```

```python
img_internal = np.zeros(img.shape)
```

```python
for i in range(len(contours)):
    if hierarchy[0][i][3] != -1: #external (en dış) contour olmayan contourlar çizilir
        cv2.drawContours(img_internal , contours ,i,255 , -1)
```

```python
plt.imshow(img_internal) #en dıştaki contourun içindeki contourlar çizildi
```
