```python
flat_chess =  cv2.imread("DATA\\flat_chessboard.png")
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
flat_chess_gray = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)

```

```python
real_chess = cv2.imread("DATA\\real_chessboard.jpg")
real_chess_gray = cv2.cvtColor(real_chess ,cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(flat_chess_gray , 64, 0.01 , 10) 
```

```python
corners
```

```python
corners = np.int0(corners)
corners
```

```python
for i in corners:
    x,y = i.ravel() # [[x, y]]  [x, y]

    cv2.circle(flat_chess ,(x,y) ,3, 255 ,-1) # keler tespit edilip boyand
    plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(real_chess_gray , 64, 0.01 , 10) 
```

```python
corners = np.int0(corners)
corners
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(real_chess , (x,y),3,255,-1)
    plt.imshow(real_chess)
```
