```python
face_cascade = cv2.CascadeClassifier("DATA\\haarcascades\\haarcascade_frontalface_default.xml") #haar cascade dosyas tanmlanr
```

```python
face_cascade
```

```python
def detect_face(img):
    
    face_img = img.copy()
    face_rect= face_cascade.detectMultiScale(face_img) #bulunan  yzlerin kordinatlar alnr
    
    for (x,y,w,h)in face_rect:
        cv2.rectangle(face_img ,(x,y),(x+w,y+h) ,(255,0,0),2)

    return face_img
```

```python
img = detect_face(img_erk)
```

```python
plt.imshow(img)
```

```python
img = detect_face(toplu_foto)
```
