```python
import numpy as np 
import cv2 
from matplotlib import pyplot as plt 
```

```python
img = cv2.imread("DATA\\sudokuu.jpg")
```

```python
plt.imshow(img , cmap="gray")
```

```python
sobelx = cv2.Sobel(img , cv2.CV_64F ,1,0 , ksize = 5) #x ynnde  gradyan
```

```python
plt.imshow(sobelx , cmap="gray")
```

```python
sobely = cv2.Sobel(img , cv2.CV_64F , 0 , 1 ,ksize = 5) # y ynnde gradyan
```

```python
plt.imshow(sobely , cmap="gray")
```

```python
laplacian = cv2.Laplacian(img , cv2.CV_64F )
```

```python
plt.imshow(laplacian )
```

```python
ret , thresh = cv2.threshold(img , 50 , 255 , cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh ,cmap = "gray")
```

```python
kernel = np.ones((4,4) ,np.uint8)
```

```python
grad =  cv2.morphologyEx(thresh , cv2.MORPH_GRADIENT , kernel)
```

```python
plt.imshow(grad)
```
